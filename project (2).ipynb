```python
import cv2
from ultralytics import YOLO

# 1. Load the pre-trained hand pose model (yolo11-pose)
# The 'n' variant is optimized for high-speed real-time performance
model = YOLO("yolo11n-pose.pt") 

# 2. Open the webcam (0 is usually the default camera)
cap = cv2.VideoCapture(0)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # 3. Run inference with tracking enabled
    # 'persist=True' maintains identity for hand movement across frames
    results = model.track(frame, persist=True, task='pose', show=False)

    # 4. Visualize the results on the frame
    # This automatically draws bounding boxes and 21 hand keypoints
    annotated_frame = results[0].plot()

    # (Optional) Access specific keypoints manually
    # Keypoints are stored as [x, y, confidence]
    if results[0].keypoints is not None:
        keypoints = results[0].keypoints.xyn.cpu().numpy() 
        # Example: print coordinates of the first detected hand's wrist (index 0)
        # print(keypoints[0][0]) 

    # 5. Display the output
    cv2.imshow("YOLO11 Real-Time Hand Tracking", annotated_frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
